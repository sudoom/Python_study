Paste your entire function, including the definition, in the box below. Do not leave any debugging print statements.

````python
def sum_digits(s):
    """ assumes s a string
        Returns an int that is the sum of all of the digits in s.
          If there are no digits in s it raises a ValueError exception. """
    x = 0
    cnt = 0
    for i in s:
        try:
            x += int(i)
        except ValueError:
            cnt += 1
            continue
    if cnt == len(s):
        raise ValueError
    else:
        return x
````

Your function should return the value `26` because the longest run of monotonically increasing integers is longer than the longest run of monotonically decreasing numbers.
* If `L = [5, 4, 10]` then the longest run of monotonically increasing numbers in `L` is `[4, 10]` and the longest run of monotonically decreasing numbers in `L` is `[5, 4]`.
````python
def longest_run(L):
    """
    Assumes L is a list of integers containing at least 2 elements.
    Finds the longest run of numbers in L, where the longest run can
    either be monotonically increasing or monotonically decreasing. 
    In case of a tie for the longest run, choose the longest run 
    that occurs first.
    Does not modify the list.
    Returns the sum of the longest run. 
    """

    dec_count = 0
    inc_count = 0
    maxcount = 0
    result = 0

    for char in range(len(L) - 1):
        if L[char] <= L[char + 1]:
            dec_count += 1
            if dec_count > maxcount:
                maxcount = dec_count
                result = char + 1
        else:
            dec_count = 0
        if L[char] >= L[char + 1]:
            inc_count += 1            
            if inc_count > maxcount:
                maxcount = inc_count
                result = char + 1
        else:
            inc_count = 0
    startposition = result - maxcount
    return sum(L[startposition:result + 1])
````

For example,

* `cipher("abcd", "dcba", "dab")` returns (order of entries in dictionary may not be the same) `({'a':'d', 'b': 'c', 'd': 'a', 'c': 'b'}, 'adc')`

````python
def cipher(map_from, map_to, code):
    key_code = {}
    for i in range(len(map_from)):
        key_code[map_from[i]] = map_to[i]
    decoded = ''
    for letter in code:
        decoded += key_code[letter]
    return (key_code, decoded)           
````

````python
class myDict(object):
    """ Implements a dictionary without using a dictionary """

    def __init__(self):
        """ initialization of your representation """
        self.keys = []
        self.values = []

    def assign(self, k, v):
        """ k (the key) and v (the value), immutable objects  """
        if k in self.keys:
            self.values[self.keys.index(k)] = v
        else:
            self.keys.append(k)
            self.values.append(v)

    def getval(self, k):
        """ k, immutable object  """
        try:
            return self.values[self.keys.index(k)]
        except:
            raise KeyError

    def delete(self, k):
        """ k, immutable object """
        try:
            self.values.pop(self.keys.index(k))
            self.keys.pop(self.keys.index(k))
        except:
            raise KeyError
````
